# Estatísticas do Grafo - Lendo de arquivo
Este notebook lê o arquivo `saida.txt` gerado automaticamente pelo `trabalho.exe` e exibe os resultados em uma tabela.

```python
import pandas as pd
import re
import os

ARQUIVO = "saida.txt"

# Verifica se o arquivo existe
if not os.path.exists(ARQUIVO):
    print(f"⚠️ Arquivo '{ARQUIVO}' não encontrado. Rode o executável primeiro para gerar as estatísticas.")
else:
    with open(ARQUIVO, "r", encoding="utf-8") as f:
        linhas = f.readlines()

    estatisticas = []
    for linha in linhas:
        if re.match(r'^\d+\. .*: ', linha):
            partes = linha.split(":", 1)
            chave = partes[0].strip()
            valor = partes[1].strip()
            estatisticas.append((chave, valor))

    df = pd.DataFrame(estatisticas, columns=["Estatística", "Valor"])

    # 👇 Força exibição de todas as linhas
    pd.set_option("display.max_rows", None)
    pd.set_option("display.max_colwidth", None)

    display(df)

```
